#### 爬師大地理系最新消息網站

```python
#引入套件
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
import time
```

```python
# 模擬使用者打開web
headers={"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36"}
```

```python
start=1 #頁數
end=10 #頁數
alltitles=[]
alldates=[]
allclass=[]
for i in range(start,end+1):
    url="https://www.geo.ntnu.edu.tw/index.php/news/page/"+str(i)+"/" # 爬師大地理系官網
    html = requests.get(url)# 取得網站連線
    soup = BeautifulSoup(html.text, 'html.parser') # 把網站變成html格式
    titles = soup.find_all('h2',class_='blog-entry-title entry-title') #搜尋標題
    dates = soup.find_all('li',class_="meta-date") #搜尋日期時間
    Classs = soup.find_all('li', class_="meta-cat") #搜尋分類
    for title in titles:
        alltitles.append(title.a.string) #取得標題
    for datetime in dates:
        alldates.append(datetime.text.strip().split()) #取得日期時間
    for Class in Classs:
        allclass.append(Class.a.string) #取得分類
    i+=1
```

```python
# 轉換成json把資料放入dict
data={}
data["日期時間"]=alldates
data['分類']=allclass
data["標題"]=alltitles
data
```

```python
#json
data_json=json.dumps(data,ensure_ascii=False) 
data_json
```

```python
# 轉換成csv
data_csv= pd.DataFrame(data) 
data_csv.head()
```

```python
#產生csv檔
data_csv.to_csv('HW3.csv') 
```
