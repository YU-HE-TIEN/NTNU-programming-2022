```python
#抓取資料
#抓取健行筆記原始碼(HTML)
import urllib.request as req
url='https://www.geo.ntnu.edu.tw/index.php/news/'
with req.urlopen(url) as r:
    data=r.read().decode('utf-8')
print(data)
```

```python
#解析資料
import bs4
root=bs4.BeautifulSoup(data,'html.parser')#html.parser表示是html資料格式
print(root.title.string)#string表示只要<title></title>之間的文字
```

```python
title=root.find('h2',class_='blog-entry-title entry-title')#透過find這個函數尋找'h2'標籤裡class='blog-entry-title entry-title'
title #只能印出一個符合條件的標籤
```

```python
title=root.find('h2',class_='blog-entry-title entry-title')
print(title.a.string)
```

```python
titles=root.find_all('h2',class_='blog-entry-title entry-title')#透過find這個函數尋找'h2'標籤裡class='blog-entry-title entry-title'
titles
```

```python
for t in titles:
    print(t.a.string)
```
